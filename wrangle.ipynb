# Acquire bedroomcnt, bathroomcnt, calculatedfinishedsquarefeet, taxvaluedollarcnt, yearbuilt, taxamount, and fips from the zillow database for all 'Single Family Residential' properties.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from env import username, password, host
```

```python

def get_db_url(db_name):
    return f'mysql+pymysql://{username}:{password}@{host}/{db_name}'


zillow=get_db_url('zillow')

```

```python
zillow=pd.read_sql('SELECT * FROM properties_2017',zillow)
```

```python
df= zillow[['bedroomcnt', 'bathroomcnt','calculatedfinishedsquarefeet', 'taxvaluedollarcnt','yearbuilt','taxamount','fips']]
```

```python
df_zillow=df.interpolate()

df_zillow
```

```python
df_zillow.isnull().sum()
```

```python
# because i used interpolate it defaults to linear method unless otherwise specified. since you need the first two 
# cells to establish the line if they are null they cannot be interpolated. from here I will add the avg

df_zillow.fillna()
```

```python
Dataset= df_zillow
X = Dataset.iloc[:,0].values


from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(X)
  
X = imputer.transform(X)
print(X)
```
